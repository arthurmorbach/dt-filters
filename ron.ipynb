nch_lvt_mac

```python
file_path = 'docs/ron_W_tran.txt'

Ws = []
rons = []

with open(file_path, 'r') as file:
    for line in file:
        # Skip comment lines
        if line.startswith(';'):
            continue
        # Split the line by comma and convert to float
        try:
            W, ron = map(float, line.split(','))
            Ws.append(W)
            rons.append(ron)
        except ValueError:
            # In case of a malformed line, print a message and skip
            print(f"Skipping malformed line: {line.strip()}")
            
# Convert lists to numpy arrays
rons = np.array(rons, dtype=float)
Ws = np.array(Ws, dtype=float)

plt.figure(1)
plt.plot(Ws*1e6, rons, label='from Virtuoso')
plt.title('nch_lvt_mac')
plt.xlabel('W (um)')
plt.ylabel('Ron (ohm)')
plt.grid(True)
plt.legend()
plt.show()
```
